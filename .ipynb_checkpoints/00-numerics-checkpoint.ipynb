## Simple startup
### Integrating a simple ode

Equation to be integrated
$$\frac{du}{dt} = -u$$

Intial condition
$$ u(0) = 1 $$

Exact analytical solution
$$u(t)=e^{-t}$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from math import exp
import matplotlib
matplotlib.rcParams.update({'font.size': 22})
```

```python
tot_time = 1.0
dt = 0.2
ntst = int(tot_time/dt)
ntrials = 5
itrial = 5

u_1,u_2,u_3,u_ex = np.zeros((4,ntst+1))
er_1, er_2, er_3 = np.zeros((3,ntrials+1))
t = np.zeros(ntst+1)

u_1[0]=1.
u_2[0]=1.
u_3[0]=1.
u_ex[0]=1.
```

```python
for i in range(1,ntst+1):
    u_1[i]=(1.0-dt)*u_1[i-1]
    u_2[i]=u_2[i-1]/(1.+dt)
    u_3[i]=u_3[i-1]*(1.0-0.5*dt)/(1+0.5*dt)
    t[i]=t[i-1]+dt
    u_ex[i] = exp(-t[i])
    

    
er_1[itrial] = abs(u_1[ntst]-u_ex[ntst])
er_2[itrial] = abs(u_2[ntst]-u_ex[ntst])
er_3[itrial] = abs(u_3[ntst]-u_ex[ntst])

```

```python
u_ex
```

```python
fig = plt.figure(1,figsize=(10,10/1.4))
plt.plot(t,u_1,'r',label='Forward Euler',lw=2.0)
plt.plot(t,u_2,'b',label='Backward Euler',lw=2.0)
plt.plot(t,u_3,'g',label='Trapezoidal',lw=2.0)
plt.plot(t,u_ex,'m',label='Exact',lw=6.0,alpha=0.5)
plt.xlabel(r'$t$'); plt.ylabel(r'$f$'); plt.legend()
plt.title('f -vs- t')
```
